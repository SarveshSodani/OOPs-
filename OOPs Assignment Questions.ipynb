1. What is Object-Oriented Programming (OOP)?

Ans. OOP is a programming paradigm based on objects and classes that helps organize code using concepts like inheritance, encapsulation, polymorphism, and abstraction.

2. What is a class in OOP?

Ans. A class is a blueprint/template used to create objects. It defines attributes variables and methods functions.

3. What is an object in OOP?

Ans. An object is an instance of a class that contains real values and can use class methods.

4. Difference between abstraction and encapsulation
Ans. Abstraction → Hides implementation details
Encapsulation → Hides data using private variables

5. What are dunder methods in Python?

Ans. Special methods with double underscores like __init__, __str__, used to define object behavior.

6. Inheritance in OOP

Ans. Inheritance allows one class (child) to use properties and methods of another class (parent).

7. What is polymorphism in OOP?

Ans.
Polymorphism means "many forms" where the same method behaves differently.
It allows flexibility by using one interface for different data types.

8. How is encapsulation achieved in Python?

Ans.
Encapsulation is achieved using private variables (with __).
Getter and setter methods are used to control access to data.

9. What is a constructor in Python?

Ans.
A constructor is a special method __init__() used to initialize objects.
It is automatically called when an object is created.

10. What are class and static methods in Python?

Ans.
A class method works with class variables and uses cls as a parameter.
A static method does not depend on class or instance variables.

11. What is method overloading in Python?

Ans.
Method overloading means defining multiple methods with the same name but different parameters.
Python does not support it directly but achieves it using default arguments.

12. What is method overriding in OOP?

Ans.
Method overriding occurs when a child class redefines a method of the parent class.
It allows specific implementation in the child class.

13. What is a property decorator in Python?

Ans.
The @property decorator is used to define getter methods.
It allows accessing methods like attributes without calling them.

14. Why is polymorphism important in OOP?

Ans.
Polymorphism increases flexibility and code reusability.
It allows the same function to work with different types of objects.

15. What is an abstract class in Python?

Ans.
An abstract class is a class that cannot be instantiated directly.
It contains abstract methods that must be implemented in child classes.

16. What are the advantages of OOP?

Ans.
OOP provides modularity, reusability, and easy maintenance of code.
It also improves security through encapsulation.

17. Difference between class variable and instance variable
Ans.
Class variables are shared among all objects of a class.
Instance variables are unique for each object.

18. What is multiple inheritance in Python?
Ans.
Multiple inheritance allows a class to inherit from more than one parent class.
It helps in combining functionalities from multiple classes.

19. Purpose of __str__ and __repr__
Ans.
__str__ provides a readable output for users.
__repr__ provides an official string representation for debugging.

20. Significance of super() function

Ans. super() is used to call methods of the parent class.
It helps in reusing code and avoiding duplication.

21. Significance of __del__ method

Ans. __del__ is a destructor method called when an object is deleted.
It is used to release resources or perform cleanup tasks.

22. Difference between @staticmethod and @classmethod
Ans. Static methods do not access class or instance variables.
Class methods can access class variables using cls.

23. How does polymorphism work with inheritance?

Ans. Polymorphism works when child classes override parent class methods.
The same method name behaves differently based on the object.

24. What is method chaining in Python OOP?

Ans. Method chaining means calling multiple methods in a single statement.
It is done by returning self from methods.

25. Purpose of __call__ method

Ans. __call__ allows an object to be called like a function.
It is useful for creating callable objects.

```python
# Q1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Bark!")

d = Dog()
d.speak()
```

```python
#Q2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        return 3.14 * 5 * 5

class Rectangle(Shape):
    def area(self):
        return 4 * 6
```

```python
#Q3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    pass

class ElectricCar(Car):
    def __init__(self, type, battery):
        super().__init__(type)
        self.battery = battery
```

```python
#Q4. Demonstrate polymorphism using Bird, Sparrow, and Penguin classes.
class Bird:
    def fly(self):
        print("Bird flies")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")
```

```python
#Q5. Demonstrate encapsulation using BankAccount class.
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def get_balance(self):
        return self.__balance
```

```python
#Q6. Demonstrate runtime polymorphism using Instrument class.
class Instrument:
    def play(self):
        pass

class Guitar(Instrument):
    def play(self):
        print("Guitar is playing")

class Piano(Instrument):
    def play(self):
        print("Piano is playing")
```

```python
#Q7. Create a class with class method and static method.
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b
```

```python
#Q8. Implement a class Person to count total persons created.
class Person:
    count = 0

    def __init__(self):
        Person.count += 1
```

```python
#Q9. Override str method in Fraction class.
class Fraction:
    def __init__(self, n, d):
        self.n = n
        self.d = d

    def __str__(self):
        return f"{self.n}/{self.d}"
```

```python
#Q10. Demonstrate operator overloading using Vector class.
class Vector:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Vector(self.x + other.x)
```

```python
#Q11. Create a Person class with greet method.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
```

```python
#Q12. Create Student class to calculate average grades.
class Student:
    def __init__(self, grades):
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)
```

```python
#Q13. Rectangle class with area calculation.
class Rectangle:
    def set_dimensions(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w
```

```python
#Q14. Employee and Manager salary calculation.
class Employee:
    def calculate_salary(self, hours, rate):
        return hours * rate

class Manager(Employee):
    def calculate_salary(self, hours, rate):
        return super().calculate_salary(hours, rate) + 5000
```

```python
#Q15. Product class with total price.
class Product:
    def __init__(self, name, price, qty):
        self.name = name
        self.price = price
        self.qty = qty

    def total_price(self):
        return self.price * self.qty
```

```python
#Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

c = Cow()
c.sound()

s = Sheep()
s.sound()
```

```python
#Q17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"{self.title} by {self.author}, published in {self.year_published}"

b = Book("Python Basics", "John Doe", 2023)
print(b.get_book_info())
```

```python
#Q18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

m = Mansion("Delhi", 50000000, 10)
print(m.address, m.price, m.number_of_rooms)
```
